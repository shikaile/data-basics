```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv('../data_sets/ZHVI.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
Focus = df[['Unnamed: 0','Michigan','Maryland', 'North Carolina','Texas', 'California', 'the District of Columbia']]
```

```python
Focus = Focus.dropna()
```

```python
Focus.rename(columns ={'Unnamed: 0': 'Dates'}, inplace = True)
```

```python
Focus['Dates'] = pd.to_datetime(Focus.Dates)
```

```python
Focus.info()
```

```python
Focus.head(3)
```

```python
clusterFocus = Focus[['Michigan','Maryland', 'North Carolina','Texas', 'California', 'the District of Columbia']]
```

```python
clusterFocus.head(3)
```

```python
kmean = KMeans(n_clusters = 7, random_state=0).fit(clusterFocus)
```

```python
kmean
```

```python
states = kmean.labels_
```

```python
len(states),clusterFocus.shape
```

```python
kmean.cluster_centers_
```

```python
clusterFocus['income-based_clusters'] = states
```

```python
clusterFocus.head(2)
```

```python
clusterFocus.groupby('income-based_clusters').mean()
```

```python
clusterFocus.groupby('income-based_clusters').median()
```
