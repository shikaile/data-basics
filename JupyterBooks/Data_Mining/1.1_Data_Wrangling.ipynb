# Important!!: transforming column info and/or reassign

```python
# no date and time in this data frame, below is only example of how to transform date object(string) into date format
pd.to_datetime(df.exampleDate)

#assigned the new date to df field name
df['exampleDate'] = pd.to_datetime(df.exampleDate)
```

```python
#selecting year, month, or day 
df.exampleDate.dt.year
df.exampleDate.dt.month
df.exampleDate.dt.day
df.exampleDate.dt.weekday
df.exampleDate.dt.weekday_name

#create new column for weekday
df['weekday_exampleDate'] = df.exampleDate.dt.weekday
```

```python
# add new column
df['newColumn']= 1
df['sampleYearDifference'] = df.yearSample2 - df.yearSample1
```
