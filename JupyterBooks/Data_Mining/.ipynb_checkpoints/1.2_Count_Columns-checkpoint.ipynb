```python
import pandas as pd
import numpy as np

#import data file

df = pd.read_csv('../data_sets/zoo2.csv')
```

```python
df.head(3)
```

```python
# Count distinct values - Descending (default)

df['hair'].value_counts()
```

```python
# Count distinct values - Ascending (default)

df['aquatic'].value_counts(ascending=True)
```

```python
# Get Relative Frequency -(Percentage %) 
# 11.6% of the total animals are venomous, while 88.4% are not

df['venomous'].value_counts(normalize=True)
```

```python
#Get specific Item count 
#counts rows=0 & rows =1 in predator column(13 predators in animal set)

df['predator'].value_counts()[0], df['predator'].value_counts()[1]
```

```python
#Count Unique Values -Size

df.groupby('animal_name').size()
```

```python
#Count Unique Values -Count

df.groupby('animal_name').count()
```

```python
# get a count of all columns across all columns

df.apply(pd.value_counts)
```
